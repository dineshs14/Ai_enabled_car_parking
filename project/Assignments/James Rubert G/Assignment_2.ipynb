## **Import libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

## **Preprocess the Dataset**

```python
#load the data set
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/naan mudhalvan/assignment 2/House Price India.csv')
```

```python
data.describe()
```

```python
data.info()
```

```python
# Drop the 'id' and 'Date' columns
data = data.drop(['id', 'Date'], axis=1)

# Convert the 'number of bedrooms' column to an integer data type
data['number of bedrooms'] = data['number of bedrooms'].astype(int)

# Normalize the dataset
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_normalized[:, :-1], data_normalized[:, -1], test_size=0.2, random_state=42)

```

## **Build the ANN Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers

# Build the model
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dense(1, activation='elu'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adamax')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=20, verbose=1)
```

## **Train and testing the model**

```python
from sklearn.metrics import mean_squared_error

# Make predictions on the testing dataset
y_pred = model.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```
