```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
train_datagen = ImageDataGenerator(rescale=1/255, zoom_range=0.2, horizontal_flip=True, vertical_flip=False)
test_datagen = ImageDataGenerator(rescale=1/255)
```

```python
x_train = train_datagen.flow_from_directory(r'F:\naan mudhalvan\assignment 3\archive\animals\animals',
                                             target_size=(64, 64),
                                             class_mode='categorical',
                                             batch_size=100)

x_test = test_datagen.flow_from_directory(r'F:\naan mudhalvan\assignment 3\archive\animals\test data',
                                           target_size=(64, 64),
                                           class_mode='categorical',
                                           batch_size=100)
```

```python
x_train.class_indices
```

```python
model = Sequential()

# input layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# flatten layer
model.add(Flatten())

# hidden layer-1
model.add(Dense(300, activation='relu'))

# hidden layer-2
model.add(Dense(150, activation='relu'))

# output layer
model.add(Dense(90, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit_generator(x_train, steps_per_epoch=len(x_train), epochs=30, validation_data=x_test, validation_steps=len(x_test)) # training the model for 50 epochs
model.save("trainedmodel_50Epoch.h5") # saving the model
```

```python
from tensorflow import keras

# Load the saved model
model = keras.models.load_model(r"F:\naan mudhalvan\assignment 3\trainedmodel_50Epoch.h5")
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(x_test, verbose=0)

# Print the test loss and accuracy
print('Test Loss:', test_loss)
print('Test Accuracy:', test_accuracy)
```
